Question 1

```python
import numpy as np
import matplotlib.pyplot as plt

N=100000
sigma=5
v=np.random.normal(0,sigma,N)

mean_power=np.mean(v**2)
variance=np.var(v)

print("Mean_power: ",mean_power)
print("Variance: ",variance)
```

From this, we can infer that Mean Power â‰ˆ Variance and the small difference here is due to finite sample size

Question 2

```python
rms_values=[1,3,5]
N=100000

plt.figure()
for rms in rms_values:
  v=np.random.normal(0,rms,N)
  plt.hist(v,bins=100,density=True,alpha=0.5,label=f'RMS={rms}')
plt.title("Histograms for different RMS Values")
plt.legend()
plt.xlabel('Voltage')
plt.ylabel('Probability density')
plt.show()
```

As we can see from the graph, with increasing RMS the histogram spreads wider, larger variance causes the graph to flatten and get broader but the mean remains centered at zero

Question 3

```python
N=100000
t=np.linspace(0,10*np.pi,N)

noise=np.random.normal(0,2,N)
sine=5*np.sin(0.1*t)
v_combined=noise+sine
plt.figure()
plt.hist(noise, bins=100, density=True,alpha=0.5, label="Noise Only")
plt.hist(v_combined, bins=100, density=True,alpha=0.5, label="Noise + Sine")
plt.title("Effect of adding slowly varying Sine Wave")
plt.legend()
plt.show()
```

As we can see form the graph, histogram becomes broader, it is no longer purely Gaussian and the  distribution spreads due to sine shifting mean slowly over time

Question 4

```python
num_ensembles=5000
samples_per_ensemble=50
true_sigma=3

variances=[]

for _ in range(num_ensembles):
  v=np.random.normal(0,true_sigma,samples_per_ensemble)
  variances.append(np.var(v))
plt.figure()
plt.hist(variances,bins=50,density=True)
plt.title("Distribution of Sample Variances")
plt.xlabel("Sample Variance")
plt.ylabel("Probability Density")
plt.show()

print("Mean of estimated variances:",np.mean(variances))
print("True variance:", true_sigma**2)
```

From the above graph, we can infer that the graph is skewed, similar to a chi-square graph, mean of sample variances is approximately similar to true variance, with increasing ensemble size the distribution becomes tighter and for a large value of N, the graph is almost gaussian
